```python
# making a basic Bokeh line graph

# importing Bookeh
from bokeh.plotting import figure
from bokeh.io import output_file, show

# prepare some data
x = [1, 2, 3, 4, 5]
y = [6, 7, 8, 9, 10]

# preepare the output file
output_file("htmls/Line.html")

# create a figure object
f = figure()

# create line plot
f.line(x,y)

# display the line plot
show(f)
```

```python
import pandas
from bokeh.plotting import figure, output_file, show

p = figure(plot_width=500, plot_height=400, tools='pan')

p.title.text = "Cool Data"
p.title.text_color = "Gray"
p.title.text_font = "times"
p.title.text_font_style = "bold"
p.xaxis.minor_tick_line_color = None
p.xaxis.axis_label = "Date"
p.yaxis.axis_label = "Intensity"

p.line([1,2,3], [4,5,6])
output_file("htmls/graph.html")

show(p)
```

```python
import pandas
from bokeh.plotting import figure, output_file, show
import pandas
from bokeh.plotting import figure
from bokeh.io import output_file, show

# read csv file
# df = pandas.read_excel("csv_files/verlegenhuken.xlsx")
df = pandas.read_excel("http://pythonhow.com/data/verlegenhuken.xlsx")

# prepare some data
temp = df["Temperature"]/10
pres = df["Pressure"] /10

# preepare the output file
output_file("htmls/Line_verlegenhuken.html")

# create a figure object
f = figure(plot_width=600, plot_height=500, tools='pan')

# create line plot
f.triangle(temp, pres, size=0.5)

f.title.text = "Teemperature and Air Pressure"
f.title.text_color = "Gray"
f.title.text_font = "times"
f.title.text_font_style = "bold"
f.xaxis.minor_tick_line_color = None
f.yaxis.minor_tick_line_color = None
f.xaxis.axis_label = "Temperature (C)"
f.yaxis.axis_label = "Pressure (hPa)"

# display the line plot
show(f)


```
